# Stackplot

```python
# if our end value is fixed then we use stack plot 

days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','k','b','r'],
             labels=['sleeping'',eating','working','playing'])


plt.xlabel('Days')
plt.ylabel('Total hrs spend')
plt.title('stack of the day time utilization')
plt.legend()
plt.show()
```

# Pie Chart

```python
slices=[7,2,8,7]
activities=['sleeping','eating','working','playing']
cols=['c','m','r','g']
plt.pie(slices,labels=activities,colors=cols,startangle=120,shadow=True,explode=(0,0.2,0,0),autopct='%1.1f%%',radius=1)
plt.title('Monday time utilization')
plt.legend()
plt.show()
```
